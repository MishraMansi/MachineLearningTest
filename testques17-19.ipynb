```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the CSV file 
df = pd.read_csv('E:\Machine Learning\ML-Test\Book1.csv')
```

```python
# 17. Calculate the average salary using NumPy
average_salary = np.mean(df["Salary"])
print(f"Average Salary: {average_salary}")
```

```python
# 18. Filter out employees with salary greater than 60,000
high_salary_employees = df[df["Salary"] > 60000]
print("\nEmployees with Salary > 60000:\n", high_salary_employees)
```

```python
# 19. Plot a bar chart representing department-wise average salary
dept_avg_salary = df.groupby("Department")["Salary"].mean()

plt.bar(dept_avg_salary.index, dept_avg_salary.values, color=['blue', 'green', 'red'])
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Average Salary by Department")
plt.xticks(rotation=45)  # Rotate department names for better visibility
plt.show()
```
