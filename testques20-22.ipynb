```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the CSV file
df = pd.read_csv('E:\Machine Learning\ML-Test\department_salary.csv')
```

```python
# 20. Plot a bar chart to visualize the number of employees in each department
dept_counts = df["Department"].value_counts()

plt.bar(dept_counts.index, dept_counts.values, color=['blue', 'purple', 'orange'])
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in Each Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# 21. Filter employees with a salary greater than 60,000
high_salary_employees = df[df["Salary"] > 60000]
print("\nEmployees with Salary > 60000:\n", high_salary_employees)
```

```python
# 22. Create a line graph for salary trends by department
dept_salary = df.groupby("Department")["Salary"].mean()

plt.plot(dept_salary.index, dept_salary.values, marker='o', linestyle='-', color='red')
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Salary Trend by Department")
plt.xticks(rotation=45)
plt.show()
```
