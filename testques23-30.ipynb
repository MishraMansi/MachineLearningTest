```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the CSV file
df = pd.read_csv('E:\Machine Learning\ML-Test\Screenshot 2025-03-28 072707.csv')
```

```python
# 23. Calculate the mean score of each subject
subject_means = df.iloc[:, 1:].mean()
print("Mean scores of each subject:\n", subject_means)
```

```python
# 24. Calculate the average score of each student
df["Average Score"] = df.iloc[:, 1:].mean(axis=1)
print("\nStudent Average Scores:\n", df[["Student Name", "Average Score"]])
```

```python
# 25. Find the student with the highest overall score
top_student = df.loc[df["Average Score"].idxmax(), "Student Name"]
print("\nTop Student:", top_student)
```

```python
# 26. Scatter plot for Math vs Science scores
plt.scatter(df["Math"], df["Science"], color='red')
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.title("Math vs Science Scores")
plt.show()
```

```python
# 27. Standardize the scores using NumPy
for subject in df.columns[1:-1]:  
    df[subject] = (df[subject] - df[subject].mean()) / df[subject].std()
print("\nStandardized Scores:\n", df)
```

```python
# 28. Bar plot for total scores of students
df["Total Score"] = df.iloc[:, 1:].sum(axis=1)
plt.bar(df["Student Name"], df["Total Score"], color=['blue', 'green', 'red', 'purple'])
plt.xlabel("Students")
plt.ylabel("Total Score")
plt.title("Total Scores of Students")
plt.xticks(rotation=45)
plt.show()
```

```python
# 29. Pie chart of subject-wise average scores
plt.pie(subject_means, labels=subject_means.index, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple'])
plt.title("Average Scores per Subject")
plt.show()
```

```python
# 30. Display students whose History score is above the average
history_avg = df["History"].mean()
above_avg_history = df[df["History"] > history_avg]
print("\nStudents with History score above average:\n", above_avg_history)
```
